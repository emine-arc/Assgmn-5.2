```python
def fizz_buzz():
    for i in range(1,100):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 5 == 0:
            print("Buzz")
        elif i % 3 == 0:
            print("Fizz")
        else:
            print("only numbers: ",i)
```

```python
fizz_buzz()
```
